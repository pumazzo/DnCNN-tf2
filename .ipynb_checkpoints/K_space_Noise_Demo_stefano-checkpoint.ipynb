```
#load lib
import numpy as np
from numpy.fft import fftshift, ifftshift, fftn, ifftn

from matplotlib import image
from matplotlib import pyplot

# tensorflow and PIL only needed for useful image conversion (jpeg to array) utilities
import tensorflow as tf
import tensorflow.keras as keras
from PIL import Image
```

```
# Fourier transform-based conversion functions

# Fourier transform from k-space to image space
def k_to_img(k):
    #undoes the shifting of the zero-frequency component to the center of the spectrum
    tmp1 = ifftshift(k, axes=range(k.ndim)) 

    #Compute the N-dimensional inverse discrete Fourier Transform of the k-space
    tmp2 = ifftn(tmp1, s=None, axes=range(k.ndim))   

    #Shift the zero-frequency component to the center of the spectrum
    img = fftshift(tmp2, axes=range(k.ndim))
    img *= np.sqrt(np.prod(np.take(k.shape, range(k.ndim))))

    return img

# Fourier transform from image space to k-space
def img_to_k(img):
    #undoes the shifting of the zero-frequency component to the center of the spectrum
    tmp1 = ifftshift(img, axes=range(img.ndim)) 

    #Compute the N-dimensional discrete Fourier Transform of the input image
    tmp2 = fftn(tmp1, s=None, axes=range(img.ndim))   

    #Shift the zero-frequency component to the center of the spectrum
    k = fftshift(tmp2, axes=range(img.ndim))
    k /= np.sqrt(np.prod(np.take(img.shape, range(img.ndim))))

    return k
```

```
# demo:
# 1) example on how transform from image space to k-space and viceversa reproducing Gioconda's image as in 
# wikipedia (https://it.wikipedia.org/wiki/K-spazio)
#
# 2) example on how to add gaussian noise to the image in k-space and convert back in the image space 
#

# read the jpeg image and convertit in numpy array format 
img = keras.preprocessing.image.load_img('Mona_Lisa_bw_square.jpeg')
img_array = keras.preprocessing.image.img_to_array(img)

# details about the image
print("PIL image info:") 
print(type(img))
print(img.format)
print(img.mode)
print(img.size)

print("NumPy array info:") 
print(type(img_array))    
print("type:",img_array.dtype)
print("shape:",img_array.shape)
```

```
# plot the image
pyplot.imshow(img)
```

```
# example on how to convert back the numpy array to the PIL image 
img_pil = keras.preprocessing.image.array_to_img(img_array, scale=False)
pyplot.imshow(img_pil)
```

```
# 1) transform img -> k-space -> plot intensity -> convert back to image space

#note: the image is grey scale but keras.preprocessing.image.img_to_array produce 
# an RGB image with same content for the 3 channels (we use only the first channel)
k = img_to_k(img_array[:,:,0])
```

```
#show content of k
print(k.shape)
print(k)
```

```
# compute intensity (absolute values of complex numbers) of k and produce an intensity representation of the k-space

k_int = np.absolute(k)
k_int_3 = np.dstack((k_int, k_int, k_int)) #recreate a 3-channel immage

# convert in a PIL RGB image for display
k_pil = keras.preprocessing.image.array_to_img(k_int_3, scale=False)
print("converting NumPy array:",type(k_pil))
pyplot.imshow(k_pil)
```

```
#save the image:
k_pil.save('k.jpg')
```

```
# convert back to image space

img_back = k_to_img(k)
img_back_3 = np.dstack((img_back, img_back, img_back))

img2_pil = keras.preprocessing.image.array_to_img(img_back_3, scale=False)
pyplot.imshow(img2_pil)
```

```
# 2) let's add complex gaussian noise to the k-space representation

mean = 0.0
scale = 1000.0 #to control the amount of noise (multiplicative factor of the complex bivariate normal distribution)
epsi = np.random.multivariate_normal(np.zeros(2), scale*0.5*np.eye(2), size=k.shape).view(np.complex) 
epsi = np.reshape(epsi, k.shape)
k_noise = k + epsi 
print(k_noise)
```

```
# convert back to image space
img_noise_back = k_to_img(k_noise)
img_noise_back_3 = np.dstack((img_noise_back, img_noise_back, img_noise_back))
```

```
img_noise_pil = keras.preprocessing.image.array_to_img(img_noise_back_3, scale=False)
pyplot.imshow(img_noise_pil)
```

```
#save image with noise
img_noise_pil.save('image_noise.jpg')
```
